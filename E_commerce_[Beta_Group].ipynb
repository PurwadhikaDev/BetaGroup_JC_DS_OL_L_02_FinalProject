**Metric Evaluation**

<img src="https://drive.google.com/uc?export=view&id=1Pn6QKBwJekcQ3vK5OCNQ3IfiNwvPW8gp" alt="Confusion Matrix Churn">

Type 1 error : False Positive  
Model memprediksi pelanggan akan churn, padahal sebenarnya tidak churn.

**Konsekuensi**: Perusahaan mengeluarkan biaya retensi yang tidak perlu, seperti pemberian diskon, cashback, kupon, atau intervensi layanan khusus kepada pelanggan yang sebenarnya masih loyal. Hal ini menyebabkan pemborosan biaya pemasaran dan sumber daya operasional.

Type 2 error : False Negative  
Model memprediksi pelanggan tidak churn, padahal sebenarnya churn.

**Konsekuensi**: Perusahaan kehilangan pelanggan yang seharusnya dapat dipertahankan. Ini berdampak pada hilangnya pendapatan jangka panjang serta meningkatnya biaya akuisisi pelanggan baru untuk menggantikan pelanggan yang churn.

**Pertimbangan Metrik:** Berdasarkan konsekuensinya, fokus utama model adalah menekan kehilangan pelanggan (meminimalkan False Negative) tanpa mengabaikan efisiensi biaya promosi (meminimalkan False Positive). Oleh karena itu, kita perlu menyeimbangkan Precision dan Recall melalui metrik F1-Score.

Namun, dalam proyek ini metrik utama yang digunakan adalah ROC-AUC untuk melihat kemampuan model dalam membedakan kelas secara keseluruhan. Meskipun ROC-AUC bisa menjadi kurang akurat pada data yang imbalanced seperti dataset ini, metrik ini tetap digunakan sebagai standar awal evaluasi performa model.

```python
def analyze_categorical(df, categorical_cols, show_percentage=True):
    plt.figure(figsize=(18, 12))
    plt.suptitle('Univariate Analysis of Categorical Columns', fontsize=18, y=1.02)

    for i, col in enumerate(categorical_cols):
        plt.subplot(3, 2, i + 1)
        ax = sns.countplot(x=col, data=df, palette='pastel')

        # Add counts and percentages on bars
        total = len(df)
        for p in ax.patches:
            count = int(p.get_height())
            if show_percentage:
                percentage = count / total * 100
                label = f'{count}\n({percentage:.1f}%)'
            else:
                label = f'{count}'
            ax.annotate(label,
                        (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='bottom', fontsize=11, color='black')

        plt.xlabel(col, fontsize=12)
        plt.ylabel('Count', fontsize=12)
        plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()
```

```python
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
analyze_categorical(df, cat_cols)
```

```python
def categorical_summary(df, categorical_cols):
    summary_list = []

    for col in categorical_cols:
        counts = df[col].value_counts()
        percentages = df[col].value_counts(normalize=True) * 100

        for cat in counts.index:
            summary_list.append({
                'Feature': col,
                'Category': cat,
                'Count': counts[cat],
                'Percentage': round(percentages[cat], 2)
            })

    summary_df = pd.DataFrame(summary_list)
    return summary_df
```

  * **Oversampling** pada kelas minoritas (misalnya menggunakan **SMOTE**)

* Penanganan yang tepat sangat penting untuk **mencegah model menjadi bias terhadap kelas mayoritas** dan memastikan performa prediksi churn yang lebih akurat.

```python
def analyze_categorical_with_target(df, categorical_cols, target='Churn', show_percentage=True):
    plt.figure(figsize=(18, 14))
    plt.suptitle(f'Categorical Features vs {target}', fontsize=18, y=1.02)

    summary_list = []

    for i, col in enumerate(categorical_cols):
        plt.subplot(3, 3, i + 1)
        ax = sns.countplot(x=col, hue=target, data=df, palette='pastel')

        total = len(df)
        for p in ax.patches:
            height = p.get_height()
            if show_percentage:
                perc = height / total * 100
                label = f'{int(height)}\n({perc:.1f}%)'
            else:
                label = f'{int(height)}'
            ax.annotate(label,
                        (p.get_x() + p.get_width() / 2., height),
                        ha='center', va='bottom', fontsize=11, color='black')

        plt.xlabel(col, fontsize=12)
        plt.ylabel('Count', fontsize=12)
        plt.xticks(rotation=45)
        plt.legend(title=target)
    plt.tight_layout()
    plt.show()
```

```python
def violin_analysis_with_mean(df, target='Churn', continuous_features=None):
    if continuous_features is None:
        continuous_features = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
        if target in continuous_features:
            continuous_features.remove(target)

    # --- Violin Plots ---
    plt.figure(figsize=(18, 30))
    plt.suptitle('Numerical/Continuous Features by ' + target, fontsize=22, y=1.02)

    for j, feature in enumerate(continuous_features):
        plt.subplot((len(continuous_features)+1)//2, 2, j+1)
        sns.violinplot(x=target, y=feature, data=df, palette=['red', 'lime'])
        plt.title(feature, fontsize=14)
        mean_val = df[feature].mean()
        plt.axhline(mean_val, linestyle='dashed', color='blue',
                    label=f"Overall Mean = {round(mean_val, 2)}")
        plt.legend(loc='best')
        plt.tight_layout()

    plt.show()

    # --- Mean Table by Target ---
    mean_by_target = df[continuous_features + [target]].groupby(target).mean().T
    mean_by_target.reset_index(inplace=True)
    mean_by_target.columns = ['Feature'] + [f"Mean_{str(col)}" for col in mean_by_target.columns[1:]]

    return mean_by_target
```

```python
# fix the outliers
def replace_outliers_with_mean(df, columns):
    for column in columns:
        # Calculate Q1 and Q3
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)

        # Calculate IQR
        IQR = Q3 - Q1

        # Define lower and upper bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Replace outliers with the mean
        mean_value = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)][column].mean()
        df.loc[(df[column] < lower_bound) | (df[column] > upper_bound), column] = mean_value

    return df
```

```python
from sklearn.preprocessing import LabelEncoder

def encode_categorical(df, cat_cols=None):

    df_encoded = df.copy()
    mappings = {}

    # Select categorical columns if not provided
    if cat_cols is None:
        cat_cols = df_encoded.select_dtypes(include=['object']).columns.tolist()

    # Encode each categorical column
    for col in cat_cols:
        le = LabelEncoder()
        df_encoded[col] = le.fit_transform(df_encoded[col])
        mappings[col] = dict(zip(le.classes_, le.transform(le.classes_)))

    return df_encoded, mappings
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(df, features=None):
    if features is None:
        # Select numeric features only
        features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

    X = df[features].copy()

    vif_data = []
    for i in range(X.shape[1]):
        vif_value = variance_inflation_factor(X.values, i)
        vif_data.append({'Feature': X.columns[i], 'VIF': round(vif_value, 2)})

    vif_df = pd.DataFrame(vif_data).sort_values(by='VIF', ascending=False).reset_index(drop=True)
    return vif_df
```

# **SHAP (Features Selection)**
**SHAP bersifat model-agnostic:**

* Nilai SHAP mengukur kontribusi setiap fitur terhadap prediksi pada **model dan instance tertentu**, dengan tetap mempertimbangkan **interaksi antar fitur**.

* Meskipun dua fitur memiliki korelasi yang tinggi, **SHAP akan membagi kontribusinya secara adil** sesuai dengan peran masing-masing fitur dalam prediksi.

* Masalah korelasi menjadi **kurang signifikan pada model berbasis tree**:

* SHAP sering digunakan bersama **XGBoost, LightGBM, atau Random Forest**, yang **tidak sensitif terhadap multikolinearitas** seperti pada model linear.

* Oleh karena itu, **menghapus fitur hanya berdasarkan nilai VIF tidaklah wajib** ketika analisis berfokus pada SHAP.

**Interpretability (Kemudahan Interpretasi):**

* SHAP mampu menyoroti **fitur mana yang benar-benar mendorong prediksi**, meskipun fitur-fitur tersebut saling berkorelasi.

* Jika dua fitur sangat berkorelasi tetapi salah satunya lebih prediktif, **SHAP akan menunjukkannya dengan jelas**, sehingga membantu Anda memfokuskan analisis pada variabel yang paling bermakna.

```python
def shap_feature_importance(model, X, top_n=15, plot=True):
    explainer = shap.Explainer(model, X)
    shap_values = explainer(X)

    # Aggregate absolute SHAP values
    shap_abs = np.abs(shap_values.values).mean(axis=0)

    # Create feature importance dataframe
    feature_importance = pd.DataFrame({
        'Feature': X.columns,
        'SHAP Importance': shap_abs
    }).sort_values(by='SHAP Importance', ascending=False)

    # Select top N features
    top_features = feature_importance.head(top_n)

    if plot:
        plt.figure(figsize=(10, 6))
        shap.summary_plot(shap_values, X, plot_type='bar', max_display=top_n, show=True)

    return top_features
```

**Penanganan Target Tidak Seimbang (Imbalanced) pada Prediksi Churn**

* Dalam dataset ini, variabel target **Churn** sangat tidak seimbang: sekitar **17% pelanggan mengalami churn**, sementara **83% pelanggan tetap bertahan**. Ketidakseimbangan ini menimbulkan tantangan dalam evaluasi model klasifikasi, karena metrik tradisional seperti **accuracy** dapat menyesatkan.  
  Sebagai contoh, model yang selalu memprediksi **“Not Churned”** tetap dapat memperoleh accuracy tinggi, meskipun gagal mendeteksi pelanggan yang benar-benar churn.

**Metrik Evaluasi untuk Data Tidak Seimbang**

**ROC-AUC (Receiver Operating Characteristic – Area Under Curve):**

* Mengukur seberapa baik model membedakan antara pelanggan yang churn dan tidak churn pada berbagai threshold.

* Namun, **ROC-AUC dapat terlihat terlalu optimistis pada dataset yang sangat imbalanced**, karena jumlah sampel kelas mayoritas yang besar dapat menaikkan nilai metrik secara artifisial.

**Precision, Recall, dan F1 Score:**

* **Precision** mengukur proporsi pelanggan yang diprediksi churn dan benar-benar churn.

* **Recall** mengukur proporsi pelanggan churn yang berhasil diidentifikasi dengan benar oleh model.

* **F1 Score** menggabungkan precision dan recall ke dalam satu metrik, sehingga memberikan ukuran performa yang lebih seimbang ketika data tidak seimbang.

* Penggunaan **F1 Score** membantu menyeimbangkan antara:
  * **False Positive** (memprediksi churn padahal pelanggan tidak churn)
  * **False Negative** (gagal memprediksi pelanggan yang benar-benar churn)  

  Hal ini sangat penting untuk menargetkan **strategi retensi pelanggan** secara efektif.

**Precision–Recall AUC (PR-AUC):**

* Merangkum kurva Precision–Recall dan **sangat berguna untuk dataset yang tidak seimbang**.

* **Berfokus pada kemampuan model dalam mengidentifikasi kelas minoritas (pelanggan churn)** tanpa terlalu dipengaruhi oleh dominasi kelas mayoritas.

```python
from sklearn.metrics import roc_auc_score, roc_curve, ConfusionMatrixDisplay, classification_report
from sklearn.model_selection import cross_val_score, StratifiedKFold

def results(model, X_train, y_pred, y_test, y_pred_train, y_train, classes, multi_class=False, show_cm=False, show_roc=False):
    if multi_class:
        accuracy = round(accuracy_score(y_pred, y_test) * 100, 2)
        train_accuracy = round(accuracy_score(y_pred_train, y_train) * 100, 2)
        precision = round(precision_score(y_pred, y_test, average='macro') * 100, 2)
        recall = round(recall_score(y_pred, y_test, average='macro') * 100, 2)
        f1_scr = round(f1_score(y_pred, y_test, average='macro') * 100, 2)
    else:
        accuracy = round(accuracy_score(y_pred, y_test) * 100, 2)
        train_accuracy = round(accuracy_score(y_pred_train, y_train) * 100, 2)
        precision = round(precision_score(y_pred, y_test) * 100, 2)
        recall = round(recall_score(y_pred, y_test) * 100, 2)
        f1_scr = round(f1_score(y_pred, y_test) * 100, 2)

    # Calculate AUC-ROC score
    auc_roc_score = roc_auc_score(y_test, y_pred)

    # Calculate AUC-ROC score using k-fold cross-validation
    kfold = StratifiedKFold(n_splits=5)
    auc_scores = cross_val_score(model, X_train, y_train, cv=kfold, scoring='roc_auc')




    print("\nTraining Accuracy: {}%".format(train_accuracy))
    print("Testing Accuracy: {}%".format(accuracy))
    print("Precision: {}%".format(precision))
    print("Recall: {}%".format(recall))
    print("F1-Score: {}%".format(f1_scr))
    # print("AUC-ROC Score: {}".format(auc_roc_score))
    # Print mean AUC-ROC score for all folds
    print("Mean AUC-ROC Score (5-fold cross-validation):", np.mean(auc_scores))
    print()

    if show_cm:
        print("Classification Report:")
        print(classification_report(y_pred, y_test, target_names=classes))
        print()
        print("Confusion Matrix:")
        fig, ax = plt.subplots(figsize=(7, 5))
        ConfusionMatrixDisplay.from_predictions(y_pred, y_test,
                                                  ax=ax,
                                                  display_labels=classes,
                                                  xticks_rotation='vertical')
        plt.show()

    if show_roc:
        # Plot ROC curve
        fpr, tpr, _ = roc_curve(y_test, y_pred)
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % auc_roc_score)
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver Operating Characteristic (ROC) Curve')
        plt.legend(loc='lower right')
        plt.show()

    return {
        'Model': model.__class__.__name__,
        'Training Accuracy': train_accuracy,
        'Testing Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1': f1_scr,
        # 'AUC-ROC Score': auc_roc_score,
        'KFold AUC-ROC Score': np.mean(auc_scores)
    }
```

```python
def train_model(model,X_train,y_train,X_test,y_test, labelEncoder, multi_class=None, show_cm = False, show_roc=False):
    # Record the start time
    start_time = time.time()

    # Fit the model on the training data
    model.fit(X_train, y_train)

    # Record the end time
    end_time = time.time()

    # Compute the training time in seconds
    training_time = end_time - start_time
    print(f"Training time: {training_time:.3f} seconds")

    # predict the model on testing data
    y_pred = model.predict(X_test)
    y_pred_train = model.predict(X_train)


    # get results
    RESULTS_LR = results(model,X_train, y_pred, y_test,y_pred_train, y_train, labelEncoder, multi_class, show_cm, show_roc)

    # Add the training time to RESULTS_LR
    RESULTS_LR['Training time (s)'] = training_time

    return RESULTS_LR, model
```

```python
def get_model_results(res):
  # Convert the dictionary to a DataFrame
  return pd.DataFrame.from_dict([res]).set_index('Model')
```

```python
MODELS = {
    'RandomForestClassifier': RandomForestClassifier(class_weight='balanced'),
    'LogisticRegression': LogisticRegression(class_weight='balanced'),
    'GaussianNB': GaussianNB(),
    'SVC': SVC(class_weight='balanced', probability=True),
    'DecisionTreeClassifier': DecisionTreeClassifier(class_weight='balanced'),
    'AdaBoostClassifier': AdaBoostClassifier(),
    'GradientBoostingClassifier': GradientBoostingClassifier(),
    'ExtraTreesClassifier': ExtraTreesClassifier(class_weight='balanced'),
    'BaggingClassifier': BaggingClassifier(),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'MLPClassifier': MLPClassifier(),
    'SGDClassifier': SGDClassifier(class_weight='balanced'),
    'XGBClassifier': XGBClassifier(),
    'LGBMClassifier': LGBMClassifier(force_col_wise=True, class_weight='balanced'),
    'CatBoostClassifier': CatBoostClassifier(silent=True),
}
```

## **ROC Curves**

```python
def plot_roc_comparison(models_dict, X_test, y_test):
    plt.figure(figsize=(12, 8))

    for name, model in models_dict.items():
        # Check if model supports predict_proba
        if hasattr(model, "predict_proba"):
            y_pred_proba = model.predict_proba(X_test)[:, 1]
        # If not, use decision_function (for SVM, etc.)
        elif hasattr(model, "decision_function"):
            y_pred_proba = model.decision_function(X_test)
        else:
            raise ValueError(f"Model {name} does not support probability prediction or decision function.")

        # Compute ROC curve and AUC
        fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
        auc_score = roc_auc_score(y_test, y_pred_proba)

        # Plot ROC curve
        plt.plot(fpr, tpr, linewidth=2, label=f"{name} (AUC={auc_score:.3f})")

    # Plot diagonal line for random classifier
    plt.plot([0, 1], [0, 1], 'k--', linewidth=1)

    # Titles and labels
    plt.title("ROC Curve Comparison of Models", fontsize=16, fontweight='bold')
    plt.xlabel("False Positive Rate (FPR)", fontsize=14)
    plt.ylabel("True Positive Rate (TPR)", fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.6)

    # Legend
    plt.legend(loc="lower right", fontsize=12)
    plt.tight_layout()
    plt.show()


# Example usage:
plot_roc_comparison(MODELS, X_test, y_test)
```
